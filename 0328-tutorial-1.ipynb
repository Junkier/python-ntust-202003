```python
### 動作流程開始
import requests
from bs4 import BeautifulSoup as bs

url = "https://www.nownews.com/cat/column/"

test = {
    "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36"
}

res = requests.get(url , headers = test )

soup = bs(res.text,"lxml")

# print(soup)
```

```python
## 抓取各文章連結

data = soup.select("div.td-animation-stack")
# print(data[0])

links = []

for ele in data:
    a_tag = ele.select("h3 a")[0]
    
    url  = a_tag["href"]
    
    links.append(url)
    
    # print(a_tag)
    # print(a_tag["href"])
    
for l in links:
    print(l)
```

```python
### 合成 function & main 

import time

def parsing_data(soup,url):
    
    # category
    category = soup.select("div.td-crumb-container div.entry-crumbs")[0].text.replace("  "," > ")

    # title 
    title = soup.select("header h1.entry-title")[0].text

    # time  
    time  = soup.select("header div.td-module-meta-info time")[0].text

    # url 

    # content 
    content = ""
    p_tags  = soup.select("div.td-post-content span[itemprop='articleBody'] p")

    for ele in p_tags:
        content += ele.text

    data = {
        "category" : category , 
        "title"    : title , 
        "time"     : time  ,
        "url"      : url ,
        "content"  : content
    }
    
    return data


def get_page_source_code(url):
    
    headers = {
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36"
    }

    res  = requests.get(url,headers=headers)
    soup = bs(res.text,"lxml")
    
    return soup 

### main 
# url  = links[1]

# soup = get_page_source_code(url)

# data = parsing_data(soup,url)

# print(data)

dataList = []

for url in links:
    
    soup = get_page_source_code(url)
    
    time.sleep(0.8)
    
    data = parsing_data(soup,url)
    
    dataList.append(data)

print("Done.")
```

```python
### 整理

import time

### def
def get_page_url(url):
    
    result = []
    
    headers = {
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36"
    }

    res = requests.get(url , headers = headers )

    soup = bs(res.text,"lxml")
    
    # 抓取各文章連結
    data = soup.select("div.td-animation-stack")

    for ele in data:
        a_tag = ele.select("h3 a")[0]
        link  = a_tag["href"]
        result.append(link)
        
    return result

def parsing_data(soup,url):
    
    # category
    category = soup.select("div.td-crumb-container div.entry-crumbs")[0].text.replace("  "," > ")

    # title 
    title = soup.select("header h1.entry-title")[0].text

    # time  
    time  = soup.select("header div.td-module-meta-info time")[0].text

    # url 

    # content 
    content = ""
    p_tags  = soup.select("div.td-post-content span[itemprop='articleBody'] p")

    for ele in p_tags:
        content += ele.text

    data = {
        "category" : category , 
        "title"    : title , 
        "time"     : time  ,
        "url"      : url ,
        "content"  : content
    }
    
    return data

def get_page_source_code(url):
    
    headers = {
        "user-agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/80.0.3987.149 Safari/537.36"
    }

    res  = requests.get(url,headers=headers)
    soup = bs(res.text,"lxml")
    
    return soup 



### main
    
## 取得各文章 url
links = []

for i in range(1,6):
    if i == 1:
        page_url = "https://www.nownews.com/cat/column/"
    else:
        page_url = "https://www.nownews.com/cat/column/page/{}/".format(i)
        
    result = get_page_url(page_url)
    
    links += result
        
    
print("取得各文章 url : OK!")
print("-"*80)
    

# 取得內文資料
dataList = []

for url in links:
    
    soup = get_page_source_code(url)
    
    time.sleep(0.8)
    
    data = parsing_data(soup,url)
    
    dataList.append(data)
    
    print("{} 爬取完畢！".format(url))
    

print("取得內文資料 : OK!")
print("-"*80)

```
